```python
from scipy.optimize import differential_evolution #импорт библиотек
import scipy.optimize as spo
import numpy as np
import math
import matplotlib.pyplot as plt
x = np.arange (1, 30, 1)
bounds = [(0 , 30)]
def f(x): #определение функции
    return (np.sin(x / 5) * np.exp(x / 10) + 5 * np.exp(-x / 2)).astype(int)
x_start = 30
result = spo.minimize(f, x_start, method = 'BFGS') #поиск минимума методом 'BFGS'
plt.plot (x, f(x)) #построение негладкой функции
plt.grid (True)
plt.show ()
print (result) #найден локальный минимум
```
