```python
import scipy.optimize as spo #импорт бибилиотек
import numpy as np
import math
import matplotlib.pyplot as plt
x = np.arange (1, 30, 1)
def f(x): #определение функции
    return np.sin(x / 5) * np.exp(x / 10) + 5 * np.exp(-x / 2)
x_start = 10
result = spo.minimize (f, x_start, method='BFGS') #поиск минимума методом 'BFGS' 
plt.plot (x, f(x)) 
plt.show ()
print (result) #минимум функции в зависимости от начальной точки x_start (в рамках задачи были рассмотрены x_start = 10, x_start = 30)
```
