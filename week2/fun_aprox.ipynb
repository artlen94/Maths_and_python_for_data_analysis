```python
import numpy as np #импорт бибилиотек
from scipy import linalg
import math 
a = np. array ([[3 , 5], [1, -1]])
b = np.array ([1, 2])
```

```python
from matplotlib import pylab as plt #исходная функция и ее график
x = np.arange (1, 16, 1)
y = np.sin(x / 5) * np.exp(x / 10) + 5 * np.exp(-x / 2)
plt.plot(x,y)
plt.show ()
```

```python
#приближение функции многочленом 1 степени
def f(x):
    return np.sin(x / 5) * np.exp(x / 10) + 5 * np.exp(-x / 2)
a = f(1)
b = f(15)
c = np.array ([[1 , 1], [1, 15]])
d = np. array ([a , b])
e = linalg.solve (c,d)
print (e)
x = np.arange (1, 16, 1)
y = e[0] + e[1]*x #система линейных уравнений многочлена 1 степени
plt.plot (x,y)
plt.show ()
```

```python
#приближение функции многочленом 3 степени
c = np.array ([[1, 1, 1, 1], [1, 4, 16, 4**3], [1, 10, 10**2, 10**3], [1, 15, 15**2, 15**3]])
d = np. array ([f(1) , f(4), f(10), f(15)])
e = linalg.solve (c,d)
print (e)
x = np.arange (1, 16, 1)
y = e[0]+e[1]*x + e[2]*x**2 + e[3]*x**3 #система линейных уравнений многочлена 1 степени
plt.plot (x,y)
plt.show ()
# многочлен 3 степени точнее апроксимирует исходную функцию
```
